# **Notebook 03: Treinamento do Modelo (Machine Learning)**

Com os dados devidamente limpos, codificados e separados, este notebook é dedicado exclusivamente ao treinamento do algoritmo. O objetivo é ajustar o modelo aos dados de treino e exportar o artefato final (o modelo treinado) para avaliação posterior, simulando um ambiente real de produção (MLOps).

### **1. Configuração e Carga dos Dados de Treino**
Vamos importar apenas os arquivos `X_train.csv` e `y_train.csv`. Os dados de teste permanecerão intocados até a fase de avaliação.

```python
import pandas as pd
from google.colab import drive

# Montando o Google Drive
drive.mount('/content/drive')

caminho_base = '/content/drive/MyDrive/Colab_Notebooks/ml_bot_detector/'

# Carregando APENAS os dados de treinamento
X_train = pd.read_csv(f'{caminho_base}X_train.csv')
y_train = pd.read_csv(f'{caminho_base}y_train.csv')

# O y_train é carregado como DataFrame, transformamos em Series para o Scikit-Learn
y_train = y_train.values.ravel()

print(f"Dados de treino carregados! X_train: {X_train.shape}, y_train: {y_train.shape}")
```

### **2. Treinamento com Random Forest**

Para este problema de classificação binária (Humano vs Bot), escolhemos o **Random Forest Classifier**.
Trata-se de um algoritmo baseado em múltiplas árvores de decisão (Ensemble), ideal para capturar as regras condicionais não-lineares que identificamos na EDA (ex: "Se tamanho for menor que 60kb E ignorar imagens, ENTÃO é Bot").

Vamos instanciar o modelo e ajustá-lo (`fit`) aos dados de treino.

```python
from sklearn.ensemble import RandomForestClassifier
import time

# Instanciando o modelo
# n_estimators=100 (100 árvores), random_state para reprodutibilidade e n_jobs=-1 para usar todos os núcleos do processador
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

print("Iniciando o treinamento do modelo. Isso pode levar alguns instantes...")
inicio = time.time()

# Fazendo o modelo aprender os padrões
modelo_rf.fit(X_train, y_train)

fim = time.time()
print(f"Treinamento concluído em {fim - inicio:.2f} segundos!")
```

### **3. Exportação do Modelo Treinado (Model Registry)**

Seguindo as boas práticas de Engenharia de Projetos em IA, não realizaremos a avaliação neste mesmo ambiente. Vamos serializar e salvar o modelo treinado (o "artefato") usando a biblioteca `joblib`.

Isso nos permite recarregar a inteligência do modelo em qualquer outro script ou servidor no futuro, sem precisar reprocessar e retreinar os dados do zero.

```python
import joblib

# Definindo o caminho para salvar o modelo treinado
caminho_modelo = f'{caminho_base}modelo_rf_bot_detector.pkl'

# Salvando o objeto do modelo no disco
joblib.dump(modelo_rf, caminho_modelo)

print(f"✅ Artefato do modelo salvo com sucesso em:\n{caminho_modelo}")
```
